```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame({
    'A':['A1','A2','A3','A4','A5'],
    'B':['B1','B2','B3','B4','B5'],
    'C':['C1','C2','C3','C4','C5'],
    'D':['D1','D2','D3','D4','D5']
})
```

```python
df1
```

```python
df2=pd.DataFrame({
    'A':['A6','A7','A8','A9','A10'],
    'B':['B6','B7','B8','B9','B10'],
    'C':['C6','C7','C8','C9','C10'],
    'D':['D6','D7','D8','D9','D10']
})
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
#merging
```

```python
left=pd.DataFrame({
    'A':['A1','A2','A3','A4','A5'],
    'B':['B1','B2','B3','B4','B5'],
    'C':['C1','C2','C3','C4','C5'],
    'D':['D1','D2','D3','D4','D5'],
    'key':['A','B','C','D','E']
})
```

```python
left
```

```python
right=pd.DataFrame({
    'A':['A6','A7','A8','A9','A10'],
    'B':['B6','B7','B8','B9','B10'],
    'C':['C6','C7','C8','C9','C10'],
    'D':['D6','D7','D8','D9','D10'],
    'key':['A','B','C','D','E']
})
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')
```

```python
data1=pd.DataFrame({
    'name':['pooja','dyanshwari','khushi','samiksha','sonam'],
    'age':[5,12,45,65,32],
    'bussiness':['Dairy','Enginner','Pharma','Doctor','Doctor'],
    'key1':[1,2,3,4,5],
    'key2':['A','B','C','D','E']
    
})
```

```python
data1
```

```python
data2=pd.DataFrame({
    'Name':['gana','gana','omkar','kamlesh','soham'],
    'age':[12,34,55,34,56],
    'bussiness':['IT','AIML','BCA','BCOM','BCS'],
    'key1':[1,2,3,4,5],
    'key2':['A','B','C','D','E']
    
})
```

```python
data2
```

```python
pd.merge(data1,data2,how='inner',on=['key1','key2'])
```

```python
pd.merge(data1,data2,how='outer',on=['key1','key2'])
```

```python
#joining
```

```python
right=pd.DataFrame({
    'A':['A1','A2','A3'],
    'B':['B1','B2','B3'],
    index:['k0','k1','k2']
})
```

```python
right
```

```python
left=pd.DataFrame({
    'C':['C1','C2','C3'],
    'D':['D1','D2','D3'],
    index:['k0','k1','k2']
})
```

```python
left
```

```python
left.join(right)
```

```python
left.join(right,how='outer')
```
