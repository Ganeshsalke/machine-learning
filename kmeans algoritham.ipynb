```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
df=make_blobs(n_samples=200,n_features=2,cluster_std=1.8,random_state=101)
```

```python
df
```

```python
df[0].shape
```

```python
df[1].shape
```

```python
plt.scatter(df[0][:,0],df[0][:,1] ,cmap='rainbow')
```

```python
#apply kmeans algoritham
```

```python
from sklearn.cluster import KMeans
```

```python
KMeans=KMeans(n_clusters=4)
```

```python
KMeans.fit(df[0])
KMeans.cluster_centers_
KMeans.labels_
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('KMeans')
ax1.scatter(df[0][:,0],df[0][:,1] ,cmap='rainbow',c=KMeans.labels_)
ax2.set_title('orignal')
ax2.scatter(df[0][:,0],df[0][:,1] ,cmap='rainbow',c=df[1])
```
